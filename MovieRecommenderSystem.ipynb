# Preprocessing

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# need to get the values of the name key in list
# but it is in form of a string
import ast
```

```python
def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L      
```

```python
movies.genres = movies.genres.apply(convert)
movies
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords']
```

```python
movies.iloc[0].cast
```

```python
def convert_cast(obj):
    l=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter +=1
        else:
            break
        
    return l
```

```python
#top 3 cast
movies['cast'] = movies['cast'].apply(convert_cast)
movies['cast']
```

```python
movies['crew'][0]
```

```python
def get_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break # there will be only one director in a movie.
        
    return l
```

```python
movies.crew = movies['crew'].apply(get_director)
```

```python
movies.head(1)
```

```python
movies.overview[0]
```

```python
movies.overview = movies.overview.apply(lambda x:x.split())
```

```python
movies.overview[0]
```

```python
movies.genres = movies.genres.apply(lambda x:[i.replace(" ","") for i in x])
movies.keywords = movies.keywords.apply(lambda x:[i.replace(" ","") for i in x])
movies.cast = movies.cast.apply(lambda x:[i.replace(" ","") for i in x])
movies.crew = movies.crew.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies.overview+movies.genres+movies.keywords+movies.cast+movies.crew
```

```python
movies.columns
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.shape
```

```python
new_df.tags = new_df.tags.apply(lambda x: x.lower())
```

```python
len(new_df.tags[0])
```

# Vectorization

```python
# Bag of Words 

from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000, stop_words='english',binary=True)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors.shape
```

```python
print(cv.get_feature_names_out())
# cosine similarity
```

# Stemming

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
new_df['tags'][0]
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
new_df['tags']
```

```python
new_df.to_csv('movies_data.csv', index=False)
```

# Similarity Check

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]]['title'])
```
